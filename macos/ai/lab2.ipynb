```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
DS = pd.read_csv('C:/Users/prate/Downloads/Dataset.csv')
print(DS)
```

```python
x = DS['month_number']
y = DS['total_profit']
plt.plot(x, y)
plt.xlabel('Month Number')
plt.ylabel('Total profit')
plt.title('Profit in a year')
plt.legend(loc='lower right', fontsize='medium')
plt.show()
```

```python
x = DS['month_number']
y = DS['total_units']
plt.plot(x, y, linewidth = 3, color = 'green', marker='o')
plt.xlabel('Month Number')
plt.ylabel('sold units number')
plt.title('units sold in a year')
plt.legend(loc='lower right', fontsize='medium')
plt.show()
```

```python
x = DS['month_number']
y1 = DS['facecream']
y2 = DS['facewash']
y3 = DS['toothpaste']
y4 = DS['bathingsoap']

plt.plot(x, y1, linewidth=3, color='green', marker='o', label='face cream')
plt.plot(x, y2, linewidth=3, color='red', marker='o', label='face wash')
plt.plot(x, y3, linewidth=3, color='blue', marker='o', label='toothpaste')
plt.plot(x, y4, linewidth=3, color='purple', marker='o', label='bathing soap')

plt.xlabel('Month Number')
plt.ylabel('Sold Units')
plt.title('Units Sold in a Year')
plt.legend(loc='lower right', fontsize='medium')
plt.show()
```

```python
x = DS['month_number']
y = DS['toothpaste']
plt.scatter(x, y, color='blue', marker='o', label='Toothpaste')
plt.xlabel('Month Number')
plt.ylabel('Sold Units of Toothpaste')
plt.title('Scatter Plot: Toothpaste Units Sold in a Year')
plt.grid('True')
plt.legend()


plt.show()
```

```python
x = ['Face Cream', 'Face Wash']
y = [DS['facecream'].sum(), DS['facewash'].sum()]
plt.bar(x, y, color=['green', 'red'])
plt.xlabel('Products')
plt.ylabel('Total Sold Units')
plt.title('Total Sold Units of Face Cream and Face Wash')
plt.show()
```

```python

x = DS['month_number']
y = DS['bathingsoap']


plt.bar(x, y, color='purple')
plt.xlabel('Month Number')
plt.ylabel('Bathing Soap Units Sold')
plt.title('Bathing Soap Units Sold Each Month')
plt.savefig('bathing_soap_sales_bar_chart.png')
plt.show()
```

```python
import matplotlib.pyplot as plt

months = DS['month_number']
total_profit = DS['total_profit']

plt.hist(total_profit)


plt.xlabel('Total Profit')
plt.ylabel('Frequency')
plt.title('Total Profit Distribution')

plt.show()
```

```python

months = DS['month_number']
bathing_soap_sales = DS['bathingsoap']
facewash_sales = DS['facewash']


fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))


ax1.plot(months, bathing_soap_sales, color='purple', marker='o')
ax1.set_xlabel('Month Number')
ax1.set_ylabel('Bathing Soap Units Sold')
ax1.set_title('Bathing Soap Sales Over Months')


ax2.plot(months, facewash_sales, color='red', marker='o')
ax2.set_xlabel('Month Number')
ax2.set_ylabel('Face Wash Units Sold')
ax2.set_title('Face Wash Sales Over Months')


plt.tight_layout()

plt.show()

```

```python
months = DS['month_number']
facecream_sales = DS['facecream']
facewash_sales = DS['facewash']
toothpaste_sales = DS['toothpaste']
bathingsoap_sales = DS['bathingsoap']
shampoo_sales = DS['shampoo']
moisturizer_sales = DS['moisturizer']


plt.stackplot(months, facecream_sales, facewash_sales, toothpaste_sales, bathingsoap_sales, shampoo_sales, moisturizer_sales,
colors=['green', 'red', 'blue', 'purple', 'black', 'yellow'])


plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.title('Product Sales Over Months')

plt.show()
```
