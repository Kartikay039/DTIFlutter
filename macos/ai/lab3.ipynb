```python
import matplotlib.pyplot as plt

exp_in_years = np.array([16, 12, 18, 4, 3, 10, 5, 12])
Performance_rating = np.array([87, 88, 89, 68, 78, 80, 75, 83])

n = exp_in_years.size
x = exp_in_years.sum()
y = Performance_rating.sum()
xy = (exp_in_years * Performance_rating).sum()
x_square = (exp_in_years * exp_in_years).sum()

slope = ((n*xy) - (x * y)) / ((n*x_square) - (x*x))
intercept = (y - slope * x) / n

predicted = (slope * exp_in_years) + intercept

plt.scatter(exp_in_years, Performance_rating, label='Data Points')
plt.plot(exp_in_years, predicted,color = 'red', label = 'regression line')
plt.xlabel('Experience in Years')
plt.ylabel('Performance Rating')
plt.legend()
plt.show()
print("predicted performance rating for 20 years experinece = ", (slope * 20) + intercept)
```
